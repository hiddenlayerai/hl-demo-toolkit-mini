# Model Scanner

Scan AI/ML models for malware, backdoors, and vulnerabilities using the HiddenLayer Python SDK.

**Prerequisites:**
- `pip install -r requirements.txt`
- `HIDDENLAYER_CLIENT_ID` and `HIDDENLAYER_CLIENT_SECRET` in `../.env`

**Self-hosted:** Add `base_url="http://your-host:port"` to the `HiddenLayer()` constructor.

**SDK Reference:** [hiddenlayer-sdk-python](https://github.com/hiddenlayerai/hiddenlayer-sdk-python)

## Setup

```python
import os
from dotenv import load_dotenv
from hiddenlayer import HiddenLayer
from hiddenlayer.lib import CommunityScanSource

load_dotenv("../.env")

client = HiddenLayer(
    client_id=os.getenv("HIDDENLAYER_CLIENT_ID"),
    client_secret=os.getenv("HIDDENLAYER_CLIENT_SECRET"),
)
```

## Scan a Local Model File

Upload and scan `test_models/safe_model.pkl` — a benign pickle. Expected result: `safe`.

```python
result = client.model_scanner.scan_file(
    model_name="demo-safe-model",
    model_path="test_models/safe_model.pkl",
    model_version="1",
    wait_for_results=True,
)

print(result)
```

## Scan a Community Model

Scan a model hosted on HuggingFace without downloading it locally.

```python
result = client.community_scanner.community_scan(
    model_name="openai/clip-vit-base-patch32",
    model_path="openai/clip-vit-base-patch32",
    model_source=CommunityScanSource.HUGGING_FACE,
    wait_for_results=True,
)

print(result)
```
